```python
from collections import UserString
import streamlit as st
import yfinance as yf 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import document
df = pd.read_excel('/Users/lizaosakue/Desktop/BAL PowerBI/BAL_DEF_updated.xlsx')
```

```python
# Aggregate the data to get the count per 'Thema'
theme_counts = df['Thema'].value_counts().reset_index()
theme_counts.columns = ['Thema', 'Totaal van Paragraaf']

# Sort the themes by the total count in descending order to match the plot
theme_counts = theme_counts.sort_values('Totaal van Paragraaf', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 8))  # Adjust the size as needed
barplot = sns.barplot(
    x='Totaal van Paragraaf', 
    y='Thema', 
    data=theme_counts,
    palette='coolwarm'  # Adjust the palette to match the original plot colors
)

# Set title and gridlines to match the Power BI plot
barplot.set_title('Aantal thema\'s per paragraaf')
plt.grid(True, which='major', linestyle='--', linewidth=0.5, color='grey')

# Show the plot
plt.show()
```
